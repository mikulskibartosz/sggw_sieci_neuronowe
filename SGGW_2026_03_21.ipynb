## Regresja liniowa

Zaimplementujemy regresję liniową od zera, używając tylko tensorów i autograd.

**Model:** $y = wx + b$

**Funkcja straty (MSE):** $L = \frac{1}{n}\sum_{i=1}^{n}(y_i - \hat{y}_i)^2$


````python
TRUE_W = 2.0
TRUE_B = 1.0

n_samples = 100
X = torch.linspace(-2, 2, n_samples).reshape(-1, 1)
X
````

````python
noise = torch.randn(n_samples, 1) * 0.3
y = TRUE_W * X + TRUE_B + noise
````

````python
plt.figure()
plt.scatter(X.numpy(), y.numpy())
plt.plot(X.numpy(), TRUE_W * X.numpy() + TRUE_B, "r-", label="Prawidłowy model")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Dane dla regresji liniowej")
plt.legend()
plt.grid(True)
plt.show()
````

````python
## Krok 1: Inicjalizacja parametrów wartościami losowymi

w = torch.randn(1, requires_grad=True)
b = torch.randn(1, requires_grad=True)

print(w)
print(b)
````

````python
## Krok 2: definicja modelu i funkcji straty
def forward(X, w, b):
  return w * X + b

def mse_loss(y_pred, y_true):
  return ((y_pred - y_true) ** 2).mean()
````

````python
## Krok 3: Pętla treningowa (gradient descent)
learning_rate = 0.1
n_epochs = 100

loss_history = []
w_history = []
b_history = []

for epoch in range(n_epochs):
  # Forward pass
  y_pred = forward(X, w, b)

  # Liczenie funkcji straty
  loss = mse_loss(y_pred, y)

  # Liczymy gradienty
  loss.backward()

  with torch.no_grad():
    w -= learning_rate * w.grad
    b -= learning_rate * b.grad

  # czyścimy gradienty
  w.grad.zero_()
  b.grad.zero_()

  loss_history.append(loss.item())
  w_history.append(w.item())
  b_history.append(b.item())

  print(f'Epoch {epoch+1}/{n_epochs}, Loss: {loss.item():.4f}')

print(f"W: Parametr modelu: {w.item()}, prawidłowy parametr: {TRUE_W}")
print(f"B: Parametr modelu: {b.item()}, prawidłowy parametr: {TRUE_B}")
````

````python
# Plot loss curve
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.plot(loss_history)
plt.xlabel('Epoch')
plt.ylabel('Loss (MSE)')
plt.title('Krzywa uczenia')
plt.grid(True)

# Plot parameter evolution
plt.subplot(1, 3, 2)
plt.plot(w_history, label='w (learned)')
plt.axhline(y=TRUE_W, color='r', linestyle='--', label=f'w (true) = {TRUE_W}')
plt.xlabel('Epoch')
plt.ylabel('w')
plt.title('Ewolucja parametru w')
plt.legend()
plt.grid(True)

plt.subplot(1, 3, 3)
plt.plot(b_history, label='b (learned)')
plt.axhline(y=TRUE_B, color='r', linestyle='--', label=f'b (true) = {TRUE_B}')
plt.xlabel('Epoch')
plt.ylabel('b')
plt.title('Ewolucja parametru b')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
````

````python
plt.figure(figsize=(10, 6))

plt.scatter(X.numpy(), y.numpy(), alpha=0.6, label='Data points')

plt.plot(X.numpy(), TRUE_W * X.numpy() + TRUE_B, 'r--', linewidth=2,
         label=f'True: y = {TRUE_W}x + {TRUE_B}')

with torch.no_grad():
    y_learned = forward(X, w, b)
plt.plot(X.numpy(), y_learned.numpy(), 'g-', linewidth=2,
         label=f'Learned: y = {w.item():.2f}x + {b.item():.2f}')

plt.xlabel('X')
plt.ylabel('y')
plt.title('Regresja liniowa - Dopasowanie modelu')
plt.legend()
plt.grid(True)
plt.savefig('linear_regression_fit.png', dpi=150, bbox_inches='tight')
plt.show()
````

## ZADANIA

````python
# ====================================
# ZADANIE A: Podstawowe operacje na tensorach
# ====================================
# Trudnosc: ⭐
# Czas: 2 minuty
#
# Instrukcje:
# 1. Stworz tensor 3x4 wypelniony losowymi liczbami z rozkladu normalnego
# 2. Zmien jego ksztalt na 2x6
# 3. Przenies tensor na GPU (jesli dostepne), a nastepnie z powrotem na CPU
# 4. Przekonwertuj na tablice NumPy i oblicz sume wszystkich elementow
#
# Wymagania:
# - Uzyj torch.randn() do stworzenia tensora
# - Uzyj .reshape() lub .view() do zmiany ksztaltu
# - Uzyj .to(device) do przenoszenia miedzy urzadzeniami
# - Koncowy wynik powinien byc pojedyncza liczba (suma)

def exercise_a():
    # TODO: Twoj kod tutaj

    # Step 1: Create 3x4 random tensor
    tensor = torch.randn(3, 4)

    # Step 2: Reshape to 2x6
    reshaped = tensor.reshape(2, 6)

    # Step 3: Move to GPU and back
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    on_device = reshaped.to(device)
    back_on_cpu = on_device.to('cpu')

    # Step 4: Convert to NumPy and sum
    numpy_array = back_on_cpu.numpy()
    total_sum = numpy_array.sum()

    return total_sum

# Test
result = exercise_a()
print(f"Sum of all elements: {result}")
````

````python
# ====================================
# ZADANIE B: Mnozenie macierzy
# ====================================
# Trudnosc: ⭐⭐
# Czas: 3-5 minut
#
# Instrukcje:
# 1. Stworz macierz A o wymiarach 3x4 (uzyj torch.arange i reshape)
# 2. Stworz macierz B o wymiarach 4x2 (uzyj torch.ones)
# 3. Oblicz iloczyn macierzy A @ B dwoma sposobami:
#    a) Uzywajac operatora @
#    b) Uzywajac torch.mm()
# 4. Sprawdz czy oba wyniki sa identyczne (uzyj torch.allclose)
#
# Wymagania:
# - Funkcja zwraca True jesli oba wyniki sa identyczne
# - Wypisz wymiary macierzy A, B i wyniku

def exercise_b():
    # TODO: Twoj kod tutaj

    # Step 1: Create matrix A (3x4)
    A = torch.arange(12, dtype=torch.float32).reshape(3, 4)

    # Step 2: Create matrix B (4x2)
    B = torch.ones(4, 2)

    # Step 3a: Matrix multiplication with @ operator
    result_operator = A @ B

    # Step 3b: Matrix multiplication with torch.mm
    result_mm = torch.mm(A, B)

    # Step 4: Check if results are identical
    are_equal = torch.allclose(result_operator, result_mm)

    return are_equal

# Test
result = exercise_b()
print(f"Results are equal: {result}")
````

````python
# ====================================
# ZADANIE C: Gradient Descent dla f(x) = x^2 - 4x + 4
# ====================================
# Trudnosc: ⭐⭐⭐
# Czas: 5-10 minut
#
# Instrukcje:
# 1. Zaimplementuj gradient descent dla funkcji f(x) = x^2 - 4x + 4
# 2. Znajdz minimum funkcji startujac od x = 10.0
# 3. Uzyj learning_rate = 0.1 i wykonaj 50 iteracji
# 4. Zapisz historie wartosci x i f(x)
# 5. Zwroc koncowa wartosc x (powinna byc bliska 2.0 - minimum funkcji)
#
# Matematycznie: f(x) = (x-2)^2, wiec minimum jest w x = 2
#
# Wymagania:
# - Uzyj autograd (requires_grad=True, backward())
# - Pamietaj o zero_grad!
# - |x_final - 2.0| < 0.01

def exercise_c():
    # TODO: Twoj kod tutaj

    # Initialize
    x = torch.tensor(10.0, requires_grad=True)
    learning_rate = 0.1
    n_iterations = 50

    x_history = []
    f_history = []

    for i in range(n_iterations):
        # Forward: compute f(x) = x^2 - 4x + 4
        f_x = x**2 - 4*x + 4

        # Backward: compute gradient
        f_x.backward()

        # Store history (before update)
        x_history.append(x.item())
        f_history.append(f_x.item())

        # Update x
        with torch.no_grad():
            x -= learning_rate * x.grad

        # Clear gradient
        x.grad.zero_()

    return x.item(), x_history, f_history

# Test
x_final, x_hist, f_hist = exercise_c()
print(f"Final x: {x_final:.4f} (expected: 2.0)")
print(f"Final f(x): {f_hist[-1]:.4f} (expected: 0.0)")
````

````python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

axes[0].plot(x_hist, 'b.-')
axes[0].axhline(y=2.0, color='r', linestyle='--', label='Minimum (x=2)')
axes[0].set_xlabel('Iteracja')
axes[0].set_ylabel('x')
axes[0].set_title('Trajektoria x w gradient descent')
axes[0].legend()
axes[0].grid(True)


axes[1].plot(f_hist, 'g.-')
axes[1].set_xlabel('Iteracja')
axes[1].set_ylabel('f(x)')
axes[1].set_title('Wartosc funkcji f(x) = x^2 - 4x + 4')
axes[1].grid(True)

plt.tight_layout()
plt.show()
````

## Wizualizacja 2D

````python
from matplotlib import cm

def loss_2d(w, b):
    return (w - 2)**2 + (b - 1)**2

w_range = np.linspace(-2, 6, 100)
b_range = np.linspace(-3, 5, 100)
W, B = np.meshgrid(w_range, b_range)
Z = loss_2d(W, B)

fig = plt.figure(figsize=(14, 5))

ax1 = fig.add_subplot(1, 2, 1, projection='3d')
surf = ax1.plot_surface(W, B, Z, cmap=cm.viridis, alpha=0.8)
ax1.set_xlabel('w')
ax1.set_ylabel('b')
ax1.set_zlabel('Loss')
ax1.set_title('Powierzchnia straty (Loss Surface)')
ax1.view_init(elev=30, azim=45)

ax2 = fig.add_subplot(1, 2, 2)
contour = ax2.contour(W, B, Z, levels=20, cmap=cm.viridis)
ax2.clabel(contour, inline=True, fontsize=8)
ax2.scatter([2], [1], color='red', s=200, marker='*', zorder=5, label='Minimum (2, 1)')
ax2.set_xlabel('w')
ax2.set_ylabel('b')
ax2.set_title('Kontury straty (widok z gory)')
ax2.legend()

plt.tight_layout()
plt.show()
````

````python
w = torch.tensor(5.0, requires_grad=True)
b = torch.tensor(4.0, requires_grad=True)
learning_rate = 0.1
n_steps = 30

w_history = [w.item()]
b_history = [b.item()]
loss_history = [loss_2d(w.item(), b.item())]

for step in range(n_steps):
    # Forward: compute loss
    loss = (w - 2)**2 + (b - 1)**2

    # Backward: compute gradients
    loss.backward()

    # Update parameters
    with torch.no_grad():
        w -= learning_rate * w.grad
        b -= learning_rate * b.grad

    # Clear gradients
    w.grad.zero_()
    b.grad.zero_()

    # Store history
    w_history.append(w.item())
    b_history.append(b.item())
    loss_history.append(loss.item())
````

````python
def run_gradient_descent(learning_rate, n_steps=50, start_x=10.0):
    x = start_x
    x_history = [x]
    f_history = [f(x)]

    for _ in range(n_steps):
        gradient = df(x)
        x = x - learning_rate * gradient
        x_history.append(x)
        f_history.append(f(x))

        if abs(x) > 1000:
            break

    return x_history, f_history

learning_rates = [0.01, 0.1, 0.5, 0.9, 1.0, 1.1]
results = {}

for lr in learning_rates:
    x_hist, f_hist = run_gradient_descent(lr, n_steps=30)
    results[lr] = {'x': x_hist, 'f': f_hist}
````

## 2.1 Co robi nn.Linear matematycznie?

**nn.Linear(in_features, out_features)** wykonuje transformację afiniczną:

$$y = xW^T + b$$

Gdzie:
- $x$ — wejście, kształt: `(batch_size, in_features)`
- $W$ — macierz wag, kształt: `(out_features, in_features)`
- $b$ — wektor bias, kształt: `(out_features,)`
- $y$ — wyjście, kształt: `(batch_size, out_features)`

**Liczba parametrów:** `in_features × out_features + out_features`

````python
class MNISTClassifier(nn.Module):
  def __init__(self):
    super().__init__()

    self.layers = nn.Sequential(
        nn.Flatten(), # (B, 1, 28, 28) -> (B, 784)
        nn.Linear(784, 256),
        nn.ReLU(),
        nn.Linear(256, 128),
        nn.ReLU(),
        nn.Linear(128, 10)
    )

  def forward(self, x):
    return self.layers(x)

model = MNISTClassifier()
print(model)
````

````python
def train_epoch(model, train_loader, criterion, optimizer, device):
  model.train()
  total_loss = 0
  correct = 0
  total = 0

  for batch_idx, (data, target) in enumerate(train_loader):
    data, target = data.to(device), target.to(device)
    output = model(data)
    loss = criterion(output, target)

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    total_loss += loss.item()
    pred = output.argmax(dim=1)
    correct += (pred == target).sum().item()
    total += target.size(0)

  return total_loss / len(train_loader), correct / total
````

````python
def evaluate(model, test_loader, criterion, device):
  model.eval()
  total_loss = 0
  correct = 0
  total = 0

  with torch.no_grad():
    for data, target in test_loader:
      data, target = data.to(device), target.to(device)

      output = model(data)
      loss = criterion(output, target)

      total_loss += loss.item()
      pred = output.argmax(dim=1)
      correct += (pred == target).sum().item()
      total += target.size(0)

  return total_loss / len(test_loader), correct / total
````

````python
## pętla treningowa

criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
n_epochs = 10

history = {
    "train_loss": [],
    "train_acc": [],
    "test_loss": [],
    "test_acc": []
}

model = model.to(device)

for epoch in range(n_epochs):
  train_loss, train_acc = train_epoch(model, train_loader, criterion, optimizer, device)
  test_loss, test_acc = evaluate(model, test_loader, criterion, device)

  history["train_loss"].append(train_loss)
  history["train_acc"].append(train_acc)
  history["test_loss"].append(test_loss)
  history["test_acc"].append(test_acc)

  print(f"Epoch {epoch+1}/{n_epochs}: "
  f"Train Loss: {train_loss:.4f}, Train Acc: {train_acc:.4f}, "
  f"Test Loss: {test_loss:.4f}, Test Acc: {test_acc:.4f}")
````

````python
# ====================================
# ZADANIE B: Zamien ReLU na Sigmoid
# ====================================
# Trudnosc: ⭐⭐
# Czas: 7 minut
#
# Instrukcje:
# 1. Stworz model z architektura 784 -> 256 -> 128 -> 10
# 2. Uzyj Sigmoid zamiast ReLU jako funkcji aktywacji
# 3. Wytrenuj przez 5 epok i porownaj z modelem ReLU
# 4. Zaobserwuj roznice w predkosci uczenia
#
# Wymagania:
# - Zamien nn.ReLU() na nn.Sigmoid()
# - Porownaj krzywe uczenia

class MNISTClassifier(nn.Module):
  def __init__(self):
    super().__init__()

    self.layers = nn.Sequential(
        nn.Flatten(), # (B, 1, 28, 28) -> (B, 784)
        nn.Linear(784, 256),
        nn.ReLU(),
        nn.Linear(256, 128),
        nn.ReLU(),
        nn.Linear(128, 10)
    )

  def forward(self, x):
    return self.layers(x)

class MLPSigmoid(nn.Module):
    """MLP with Sigmoid activation instead of ReLU."""

    def __init__(self):
        super().__init__()
        self.layers = nn.Sequential(
            nn.Flatten(),
            nn.Linear(784, 256),
            nn.Sigmoid(),
            nn.Linear(256, 128),
            nn.Sigmoid(),
            nn.Linear(128, 10)
        )

    def forward(self, x):
        return self.layers(x)

model_relu = MNISTClassifier().to(device)
model_sigmoid = MLPSigmoid().to(device)

opt_relu = optim.Adam(model_relu.parameters(), lr=0.001)
opt_sigmoid = optim.Adam(model_sigmoid.parameters(), lr=0.001)

relu_history = []
sigmoid_history = []

for epoch in range(5):
    _, relu_acc = train_epoch(model_relu, train_loader, criterion, opt_relu, device)
    _, sigmoid_acc = train_epoch(model_sigmoid, train_loader, criterion, opt_sigmoid, device)

    relu_history.append(relu_acc)
    sigmoid_history.append(sigmoid_acc)

    print(f"{epoch+1:<8}{relu_acc:<12.2f}{sigmoid_acc:<12.2f}")
````

````python
# ====================================
# ZADANIE C: Dodaj Dropout
# ====================================
# Trudnosc: ⭐⭐⭐
# Czas: 10 minut
#
# Instrukcje:
# 1. Dodaj Dropout(0.5) po kazdym ReLU w modelu 784 -> 256 -> 128 -> 10
# 2. Wytrenuj przez 10 epok
# 3. Porownaj train vs test accuracy (overfitting?)
# 4. Zaobserwuj czy Dropout pomaga w regularyzacji
#
# Wymagania:
# - Dodaj nn.Dropout(0.5) po kazdym nn.ReLU()
# - Nie dodawaj Dropout po ostatniej warstwie!
# - Porownaj z modelem bez Dropout

class MLPDropout(nn.Module):
    """MLP with Dropout regularization."""

    def __init__(self, dropout_rate=0.5):
        super().__init__()
        self.layers = nn.Sequential(
            nn.Flatten(),
            nn.Linear(784, 256),
            nn.ReLU(),
            nn.Dropout(dropout_rate),
            nn.Linear(256, 128),
            nn.ReLU(),
            nn.Dropout(dropout_rate),
            nn.Linear(128, 10)
        )

    def forward(self, x):
        return self.layers(x)

model_no_dropout = MNISTClassifier().to(device)
model_with_dropout = MLPDropout(dropout_rate=0.5).to(device)

opt_no_drop = optim.Adam(model_no_dropout.parameters(), lr=0.001)
opt_with_drop = optim.Adam(model_with_dropout.parameters(), lr=0.001)

history_no_drop = {'train': [], 'test': []}
history_with_drop = {'train': [], 'test': []}

for epoch in range(10):
    print(f"{epoch+1:<8}", end='')
    _, train_acc_nd = train_epoch(model_no_dropout, train_loader, criterion, opt_no_drop, device)
    _, test_acc_nd = evaluate(model_no_dropout, test_loader, criterion, device)

    _, train_acc_wd = train_epoch(model_with_dropout, train_loader, criterion, opt_with_drop, device)
    _, test_acc_wd = evaluate(model_with_dropout, test_loader, criterion, device)

    history_no_drop['train'].append(train_acc_nd)
    history_no_drop['test'].append(test_acc_nd)
    history_with_drop['train'].append(train_acc_wd)
    history_with_drop['test'].append(test_acc_wd)

    print(f"{epoch+1:<8}{train_acc_nd:<14.2f}{test_acc_nd:<14.2f}{train_acc_wd:<14.2f}{test_acc_wd:<14.2f}")
````

````python
def manual_conv2d(image, kernel):
    h, w = image.shape
    kh, kw = kernel.shape

    out_h = h - kh + 1
    out_w = w - kw + 1

    output = np.zeros((out_h, out_w))

    for i in range(out_h):
        for j in range(out_w):
            patch = image[i:i+kh, j:j+kw]
            output[i, j] = np.sum(patch * kernel)

    return output
````

Formuła wymiarów wyjściowych

$$\text{output\_size} = \left\lfloor \frac{\text{input\_size} - \text{kernel\_size} + 2 \times \text{padding}}{\text{stride}} \right\rfloor + 1$$

**Przykład:** input=32, kernel=3, padding=1, stride=1
$$\text{output} = \frac{32 - 3 + 2 \times 1}{1} + 1 = \frac{31}{1} + 1 = 32$$

**Praktyczna zasada:** `padding = kernel_size // 2` zachowuje wymiary (dla stride=1).

````python
def conv_output_size(input_size, kernel_size, padding=0, stride=1):
    return (input_size - kernel_size + 2 * padding) // stride + 1
````

````python
class CNNModel(nn.Module):
    def __init__(self):
        super(CNNModel, self).__init__()

        self.layers = nn.Sequential(
            # Block 1: Conv + funkcja aktywacji + maxpool
            nn.Conv2d(3, 32, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2, stride=2), # 32x32 -> 16x16

            # Block 2: Conv + funkcja aktywacji + maxpool
            nn.Conv2d(32, 64, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2, stride=2), # 16x16 -> 8x8

            nn.Flatten(),
            nn.Linear(64 * 8 * 8, 256),
            nn.ReLU(),
            nn.Linear(256, 10)
        )
    def forward(self, x):
        return self.layers(x)
````

````python
model = CNNModel().to(device)
print(model)
````

````python
def train_epoch(model, train_loader, criterion, optimizer, device):
    """Train for one epoch."""
    model.train()
    total_loss = 0
    correct = 0
    total = 0

    for data, target in train_loader:
        data, target = data.to(device), target.to(device)

        optimizer.zero_grad()
        output = model(data)
        loss = criterion(output, target)
        loss.backward()
        optimizer.step()

        total_loss += loss.item()
        pred = output.argmax(dim=1)
        correct += (pred == target).sum().item()
        total += target.size(0)

    return total_loss / len(train_loader), 100. * correct / total


def evaluate(model, test_loader, criterion, device):
    """Evaluate on test set."""
    model.eval()
    total_loss = 0
    correct = 0
    total = 0

    with torch.no_grad():
        for data, target in test_loader:
            data, target = data.to(device), target.to(device)
            output = model(data)
            loss = criterion(output, target)

            total_loss += loss.item()
            pred = output.argmax(dim=1)
            correct += (pred == target).sum().item()
            total += target.size(0)

    return total_loss / len(test_loader), 100. * correct / total
````

````python
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
n_epochs = 15

history = {'train_loss': [], 'train_acc': [], 'test_loss': [], 'test_acc': []}

for epoch in range(n_epochs):
    train_loss, train_acc = train_epoch(model, train_loader, criterion, optimizer, device)
    test_loss, test_acc = evaluate(model, test_loader, criterion, device)

    history['train_loss'].append(train_loss)
    history['train_acc'].append(train_acc)
    history['test_loss'].append(test_loss)
    history['test_acc'].append(test_acc)

    print(f"{epoch+1:<8}{train_loss:<12.4f}{train_acc:<12.2f}{test_loss:<12.4f}{test_acc:<12.2f}")
````

````python
def clean_text(text):
    text = text.lower()
    text = re.sub(r'<[^>]+>', ' ', text)
    text = re.sub(r'[^a-z\s]', '', text)
    text = ' '.join(text.split())
    return text

def tokenize(text):
    return text.split()
````

````python
def build_vocab(texts, max_vocab=10000, min_freq=2):
    counter = Counter()
    for text in texts:
        tokens = tokenize(clean_text(text))
        counter.update(tokens)

    filtered = [(word, count) for word, count in counter.most_common()
                if count >= min_freq]
    filtered = filtered[:max_vocab - 2]

    vocab = {'<pad>': 0, '<unk>': 1}
    for idx, (word, _) in enumerate(filtered):
        vocab[word] = idx + 2

    return vocab

sample_texts = [
    "This movie was great!",
    "I loved this film!",
    "The movie was terrible.",
    "I hated this film."
]

vocab = build_vocab(sample_texts, max_vocab=100, min_freq=1)
len(vocab)
````

````python
def text_to_tensor(text, vocab, max_len=256):
    tokens = tokenize(clean_text(text))

    unk_idx = vocab['<unk>']
    indices = [vocab.get(token, unk_idx) for token in tokens]

    pad_idx = vocab['<pad>']
    if len(indices) < max_len:
        indices = indices + [pad_idx] * (max_len - len(indices))
    else:
        indices = indices[:max_len]

    return torch.tensor(indices, dtype=torch.long)
````

````python
def download_imdb(data_dir='./data/imdb'):
    url = 'https://ai.stanford.edu/~amaas/data/sentiment/aclImdb_v1.tar.gz'
    filename = os.path.join(data_dir, 'aclImdb_v1.tar.gz')

    if not os.path.exists(data_dir):
        os.makedirs(data_dir)

    if not os.path.exists(os.path.join(data_dir, 'aclImdb')):
        urllib.request.urlretrieve(url, filename)
        print(f"Pobrano plik: {filename}")
        with tarfile.open(filename, 'r:gz') as tar:
            tar.extractall(data_dir)

    print("Pobrane dane IMDB.")
    return os.path.join(data_dir, 'aclImdb')

imdb_path = download_imdb()
````

````python
def load_imdb_data(imdb_path, subset='train', max_samples=None):
    texts = []
    labels = []

    for label_type, label_value in [('neg', 0), ('pos', 1)]:
        folder = os.path.join(imdb_path, subset, label_type)
        files = os.listdir(folder)

        if max_samples:
            files = files[:max_samples]

        for filename in files:
            with open(os.path.join(folder, filename), 'r', encoding='utf-8') as f:
                texts.append(f.read())
                labels.append(label_value)

    return texts, labels

train_texts, train_labels = load_imdb_data(imdb_path, 'train', max_samples=5000)
test_texts, test_labels = load_imdb_data(imdb_path, 'test', max_samples=2500)
````

````python
class IMDbDataset(Dataset):

    def __init__(self, texts, labels, vocab, max_len=256):
        self.texts = texts
        self.labels = labels
        self.vocab = vocab
        self.max_len = max_len

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        text = self.texts[idx]
        label = self.labels[idx]
        tensor = text_to_tensor(text, self.vocab, self.max_len)

        return tensor, torch.tensor(label, dtype=torch.float32)
````

````python
class SentimentLSTM(nn.Module):
    def __init__(self, vocab_size, embed_dim=100, hidden_dim=128,
                 num_layers=1, dropout=0.5):
        super().__init__()

        self.hidden_dim = hidden_dim
        self.num_layers = num_layers

        # padding_idx!
        self.embedding = nn.Embedding(vocab_size, embed_dim, padding_idx=0)


        self.lstm = nn.LSTM(
            input_size=embed_dim,
            hidden_size=hidden_dim,
            num_layers=num_layers,
            batch_first=True,
            bidirectional=True,
            dropout=dropout if num_layers > 1 else 0
        )

        self.dropout = nn.Dropout(dropout)

        # hidden_dim * 2 <- bidirectional
        self.fc = nn.Linear(hidden_dim * 2, 1)

    def forward(self, x):
        embedded = self.embedding(x)
        embedded = self.dropout(embedded)

        output, (hidden, cell) = self.lstm(embedded)

        hidden_concat = torch.cat((hidden[-2], hidden[-1]), dim=1)

        hidden_concat = self.dropout(hidden_concat)
        output = self.fc(hidden_concat)

        return output.squeeze(1)

model = SentimentLSTM(
    vocab_size=vocab_size,
    embed_dim=100,
    hidden_dim=128,
    num_layers=2,
    dropout=0.5
).to(device)

print(model)
````

````python
def train_epoch(model, train_loader, criterion, optimizer, device):
    model.train()
    total_loss = 0
    correct = 0
    total = 0

    for batch_text, batch_labels in train_loader:
        batch_text = batch_text.to(device)
        batch_labels = batch_labels.to(device)

        optimizer.zero_grad()
        outputs = model(batch_text)
        loss = criterion(outputs, batch_labels)

        loss.backward()

        torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)

        optimizer.step()

        total_loss += loss.item()
        predictions = (torch.sigmoid(outputs) > 0.5).float()
        correct += (predictions == batch_labels).sum().item()
        total += batch_labels.size(0)

    avg_loss = total_loss / len(train_loader)
    accuracy = 100. * correct / total

    return avg_loss, accuracy


def evaluate(model, test_loader, criterion, device):
    model.eval()
    total_loss = 0
    correct = 0
    total = 0
    all_preds = []
    all_labels = []

    with torch.no_grad():
        for batch_text, batch_labels in test_loader:
            batch_text = batch_text.to(device)
            batch_labels = batch_labels.to(device)

            outputs = model(batch_text)
            loss = criterion(outputs, batch_labels)

            total_loss += loss.item()
            predictions = (torch.sigmoid(outputs) > 0.5).float()
            correct += (predictions == batch_labels).sum().item()
            total += batch_labels.size(0)

            all_preds.extend(predictions.cpu().numpy())
            all_labels.extend(batch_labels.cpu().numpy())

    avg_loss = total_loss / len(test_loader)
    accuracy = 100. * correct / total

    return avg_loss, accuracy, all_preds, all_labels
````

````python
criterion = nn.BCEWithLogitsLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
N_EPOCHS = 10

for epoch in range(N_EPOCHS):
    train_loss, train_acc = train_epoch(model, train_loader, criterion, optimizer, device)
    test_loss, test_acc, _, _ = evaluate(model, test_loader, criterion, device)

    history['train_loss'].append(train_loss)
    history['train_acc'].append(train_acc)
    history['test_loss'].append(test_loss)
    history['test_acc'].append(test_acc)

    print(f"{epoch+1:<8}{train_loss:<12.4f}{train_acc:<12.2f}{test_loss:<12.4f}{test_acc:<12.2f}")
````

````python
# ====================================
# ZADANIE A: Zamien LSTM na RNN
# ====================================
# Trudnosc: ⭐
# Czas: 5 minut
#
# Instrukcje:
# 1. Skopiuj klase SentimentLSTM i zmien na SentimentRNN
# 2. Zamien nn.LSTM na nn.RNN
# 3. Uwaga: RNN nie ma cell state, tylko hidden state!
# 4. Wytrenuj przez 5 epok i porownaj accuracy z LSTM
#
# Wymagania:
# - Zamien self.lstm na self.rnn
# - W forward() zmien output, (hidden, cell) na output, hidden

class SentimentRNN(nn.Module):

    def __init__(self, vocab_size, embed_dim=100, hidden_dim=128, num_layers=1, dropout=0.5):
        super().__init__()

        self.embedding = nn.Embedding(vocab_size, embed_dim, padding_idx=0)

        self.rnn = nn.RNN(
            input_size=embed_dim,
            hidden_size=hidden_dim,
            num_layers=num_layers,
            batch_first=True,
            bidirectional=True,
            dropout=dropout if num_layers > 1 else 0
        )

        self.dropout = nn.Dropout(dropout)
        self.fc = nn.Linear(hidden_dim * 2, 1)

    def forward(self, x):
        embedded = self.embedding(x)
        embedded = self.dropout(embedded)

        output, hidden = self.rnn(embedded)

        hidden_concat = torch.cat((hidden[-2], hidden[-1]), dim=1)
        hidden_concat = self.dropout(hidden_concat)
        output = self.fc(hidden_concat)

        return output.squeeze(1)

# Test
model_rnn = SentimentRNN(vocab_size, hidden_dim=64, num_layers=1).to(device)
optimizer_rnn = optim.Adam(model_rnn.parameters(), lr=0.001)
for epoch in range(5):
    train_loss, train_acc = train_epoch(model_rnn, train_loader, criterion, optimizer_rnn, device)
    test_loss, test_acc, _, _ = evaluate(model_rnn, test_loader, criterion, device)
    print(f"Epoch {epoch+1}: Train={train_acc:.2f}%, Test={test_acc:.2f}%")
````

````python
# ====================================
# ZADANIE B: Eksperymentuj z dropout
# ====================================
# Trudnosc: ⭐⭐
# Czas: 10 minut
#
# Instrukcje:
# 1. Wytrenuj modele z roznymi wartosciami dropout: 0.0, 0.3, 0.5, 0.7
# 2. Dla kazdego wykonaj 5 epok treningu
# 3. Porownaj test accuracy
# 4. Narysuj wykres: dropout vs accuracy
#
# Wymagania:
# - Uzyj tej samej architektury SentimentLSTM
# - Zapisz train i test accuracy dla kazdego dropout

dropout_values = [0.0, 0.3, 0.5, 0.7]
results = []

for dropout in dropout_values:
    model_drop = SentimentLSTM(
        vocab_size, embed_dim=100, hidden_dim=64,
        num_layers=1, dropout=dropout
    ).to(device)

    optimizer_drop = optim.Adam(model_drop.parameters(), lr=0.001)

    for epoch in range(5):
        train_epoch(model_drop, train_loader, criterion, optimizer_drop, device)

    _, train_acc = train_epoch(model_drop, train_loader, criterion, optimizer_drop, device)
    _, test_acc, _, _ = evaluate(model_drop, test_loader, criterion, device)

    results.append((dropout, train_acc, test_acc))
    print(f"{dropout:<10}{train_acc:<12.2f}{test_acc:<12.2f}")
````
